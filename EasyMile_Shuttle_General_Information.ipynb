```python
def find_between( s, first, last ):
    try:
        start = s.index( first ) + len( first )
        end = s.index( last, start )
        return s[start:end]
    except ValueError:
        return ""

##Process data from API/vehicle_states
data_states = states.text
data_states = re.sub('"', '', data_states)
vehicle_id = find_between(data_states,"vehicle_id:",",theta")
orientation = float(find_between(data_states,"theta:",",speed"))
speed = float(find_between(data_states,"speed:",",ramp"))
ramp_status = find_between(data_states,"ramp:",",platform")
payload = find_between(data_states,"payload:",",number_of_passengers")
no_of_passengers = find_between(data_states,"number_of_passengers:",",mode")
pos_long = float(find_between(data_states,"lon:",",lat"))
pos_lat = float(find_between(data_states,"lat:",",last_seen"))
last_seen = find_between(data_states,"last_seen:",",interior_temperature")
door_status = find_between(data_states,"door:",",disengagement_reason")
connection_status = find_between(data_states,"connection:",",battery")
battery = float(find_between(data_states,"battery:",",activity"))
progress_to_next_stop = float(find_between(data_states,"progress:",",next_stop_index"))
next_stop_index = float(find_between(data_states,"next_stop_index:","},line_name"))
line_name = find_between(data_states,"line_name:",",line_id")
eta_i4 = find_between(data_states,"innovation 4.0,eta:","},{")
eta_oppHSSML = find_between(data_states,"Opp HSSML,eta:","},{")
eta_29HMKT = find_between(data_states,"29 HMK Terrace,eta:","},{")
eta_BIZ2 = find_between(data_states,"BIZ 2,eta:","},{")
eta_i4_2 = find_between(data_states,"BIZ 2","]")
eta_i4_2 = find_between(eta_i4_2, "innovation 4.0,eta:", "}")

##Process data from API/lines
data_lines = lines.text
data_lines = re.sub('"', '', data_lines)
if line_name == "Full Route":
    data_lines = find_between(data_lines,"27598842-b811-11e9-8a06-0a5864607f03","]")
elif line_name == "Demo Route":
    data_lines = find_between(data_lines,"3ca5a83e-b811-11e9-8cd1-0a5864607f03","]")
elif line_name == "All stops":
    data_lines = find_between(data_lines,"613a4c7a-b7f0-11e9-aa3d-0a5864607f02","]")
else:
    raise Exception('line_name is not valid. It should be Full Route / Demo Route / All stops')
data_lines = data_lines.split('station_id:')

stop_name = []
leave_aft_prd = []
close_door_aft_prd = []
for i in data_lines:
    if i[0].isdigit():
        data_lines_split = i.split(',')
        if int(data_lines_split[0]) == 0:
            stop_name.append('Innovation 4.0')
        elif int(data_lines_split[0]) == 11:
            stop_name.append('Opp HSSML')
        elif int(data_lines_split[0]) == 1:
            stop_name.append('29 HMK Terrace')
        elif int(data_lines_split[0]) == 12:
            stop_name.append('BIZ 2')
        else:
            raise Exception('Stop ID is not defined')
        
        split_again = data_lines_split[1].split(':')
        leave_aft_prd.append(split_again[1])
        split_again = data_lines_split[2].split(':')
        split_again[1] = split_again[1][:-1]
        close_door_aft_prd.append(split_again[1])    
df = pd.DataFrame({'Stop name':stop_name, 'Leave after (s)': leave_aft_prd, 'close door after (s)': close_door_aft_prd})

##Process data from API/site
data_site = site.text
data_site = re.sub('"', '', data_site)
data_site_stations = find_between(data_site,"stations:","}}]}")
data_site_intersections = find_between(data_site,"intersections:","}}]}")
data_site_edges = find_between(data_site,"edges:","}}]}")

stations_coordinates = []
stations_long = []
stations_lat = []
intersections_coordinates = []
intersections_long = []
intersections_lat = []
edges_coordinates = []
edges_long = []
edges_lat = []
edges_split = []
route_distance = 0;
while "coordinates:[" in data_site_stations:
    stations_coordinates.append(find_between(data_site_stations,"coordinates:[","]"))
    data_site_stations = data_site_stations.split('coordinates:[',1)[1]
for i in stations_coordinates:
    stations_long.append(float(i.split(',')[0]))
    stations_lat.append(float(i.split(',')[1]))
while "coordinates:[" in data_site_intersections:
    intersections_coordinates.append(find_between(data_site_intersections,"coordinates:[","]"))
    data_site_intersections = data_site_intersections.split('coordinates:[',1)[1]
for i in intersections_coordinates:
    intersections_long.append(float(i.split(',')[0]))
    intersections_lat.append(float(i.split(',')[1]))
edges_split = data_site_edges.split('length:')
for i in edges_split:
    if i[0].isdigit():
        route_distance += float(i.split(',',1)[0])
        split_again = i.split('coordinates:[',1)[1]
        while '[' in split_again:
            edges_coordinates.append(find_between(split_again,"[","]"))
            split_again = split_again.split(']',1)[1]
for i in edges_coordinates:
    edges_long.append(float(i.split(',')[0]))
    edges_lat.append(float(i.split(',')[1]))
```

```python
##Display information
print('Vehicle Information:')
print('\t vehicle ID number: ', vehicle_id)
print('\t last seen time: ', last_seen)
print('\t current position (on map): ', pos_long, ', ', pos_lat)
print('\t current orientation: ', orientation)
print('\t current speed: ', speed)
print('\t no. of passengers: ', no_of_passengers)
print('\t payload: ', payload)
print('\t battery: ', battery)
print('\t connection status: ', connection_status)
print('\t door status: ', door_status)
print('\t ramp status: ', ramp_status, '\n')
print('Route Information:')
print('\t operation mode: ', line_name)
if next_stop_index == 1:
    print('\t next stop: Opp HSSML')
elif next_stop_index == 2:
    print('\t next stop: 29 HMK Terrace')
elif next_stop_index == 3:
    print('\t next stop: BIZ 2')
else:
    print('\t next stop: Innovation 4.0')
print('\t progress to next stop: ', progress_to_next_stop)
print('\t eta to Innovation4.0 (depart): ', eta_i4)
print('\t eta to Opp HSSML: ', eta_oppHSSML)
print('\t eta to 29 HMK Terrace: ', eta_29HMKT)
print('\t eta to BIZ 2: ', eta_BIZ2)
print('\t eta to Innovation4.0 (return): ', eta_i4_2)
print('\t full route distance: ', route_distance, '\n')
print(df, '\n')
print('Map Legend: \n Purple square: current location \n Red dots: stops \n Green dots: road junctions \n Yellow triangles: route waypoints')
fig = plt.figure()
plt.plot(stations_long, stations_lat, 'ro')
plt.plot(intersections_long, intersections_lat, 'go')
plt.plot(edges_long, edges_lat, 'y1')
plt.plot(pos_long, pos_lat, 'ms')
mplleaflet.display(fig=fig)
```
